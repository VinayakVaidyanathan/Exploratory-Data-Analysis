## Data Analysis

## 1. Uni-Variate Analysis
(Analysis Involving Single Variable)

```python
fig, axs= plt.subplots(nrows=2,ncols=2,figsize=(10,10))

sns.countplot(df['Segment'],ax=axs[0][0])
sns.countplot(df['Category'],ax=axs[0][1])
sns.countplot(df['Region'],ax=axs[1][0])
sns.countplot(df['Ship Mode'],ax=axs[1][1])

axs[0][0].set_title('Segment')
axs[0][1].set_title('Category')
axs[1][0].set_title('Region')
axs[1][1].set_title('Ship Mode')
```

We can see the nature of demand for our products by Segment,Category,Region and Ship Mode

Binders is the most popular product amongst customers

```python
plt.figure(figsize=(20,10))
sns.countplot(df['State'])
plt.xticks(rotation=90)
plt.title('State',fontsize=25)
```

We can see that California accounts for the majority of our customer-base.

```python
plt.figure(figsize=(20,10))
sns.countplot(df['Quantity'])
plt.title('Quantity',fontsize=25)
```

```python
plt.figure(figsize=(20,10))
sns.countplot(df['Discount'])
plt.title('Discount',fontsize=25)
```

```python
fig, axs= plt.subplots(nrows=2,ncols=2,figsize=(10,10))

sns.distplot(df['Sales'],color='green',ax=axs[0][0])
sns.distplot(df['Profit'],color='blue',ax=axs[0][1])
sns.distplot(df['Discount'],color='red',ax=axs[1][0])
sns.distplot(df['Quantity'],color='orange',ax=axs[1][1])

axs[0][0].set_title('Sales Distribution',fontsize=10)
axs[0][1].set_title('Profit Distribution',fontsize=10)
axs[1][0].set_title('Discount Distribution',fontsize=10)
axs[1][1].set_title('Quantity Distribution',fontsize=10)
```

## 3. Correlation Analysis

Checking how change in each variable effects the others

```python
fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(np.round(df.corr(),2), annot=True,ax=ax)
```

As we can see Profit and Sales have a positive correlation of 48% whereas Discount and Profit have a negative correlation of 22%. The rest of the variables are fairly independent of each other.

## 4. Analysis of Profit and Sales

## Based on Category

```python
dfp1 = (df['Category'].value_counts()/len(df['Category']))*100
dfp1
```

We can see Office supplies make up about 60.3% of our orders, furniture 21.22% and Technology 18.48%

```python
dfp1.plot(kind='bar')
plt.xlabel('Category')
plt.ylabel('No of Orders')
plt.show()
```

```python
df_c = df.groupby(['Category'])[['Profit','Sales','Discount']].mean()
df_c
```

```python
df_c.plot(kind='bar')
```

```python
df_c.plot.pie(subplots=True,
            autopct='%1.1f%%',
            figsize=(18,20),
            startangle= 90,
            shadow=True,
            labels=df_c.index)
plt.title("Category-wise Analysis of Profit,Sales and Discount")
```

Technology products are the best selling and most profitable category accounting for 73% of our profits. It is interesting that even though maximum orders are placed for office supplies they make the least amount of profit.

## Conclusions

- The business is healthy as profits and sales are high in almost all segments. However, there is room for improvement. 
- We can see the moderately negative correlation between profits and discounts. Though correlation does not imply causation, discounts should be cut in products  and regions high in sales but low in profits.
- Technology products have the highest sales.
- Discounts should be cut in Home Office Segment.
